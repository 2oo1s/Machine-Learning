```python
import pandas as pd

housing = pd.read_csv('/content/sample_data/housing.csv')
housing.head()  # 데이터 처음 다섯행 확인
```

```python
# 데이터에 대한 간략한 설명
# 전체 행 수, 각 특성의 데이터 타입과 널이 아닌 값의 개수를 확인하는 데 유용
housing.info()
```

```python
housing["ocean_proximity"].value_counts()
```

```python
housing.hist(bins=50, figsize=(20,15), color='orange', alpha=0.5)
# hist() 옵션별 기능
# 1. bins : 데이터를 몇 개의 구간으로 쪼개어 히스토그램을 그릴지 설정
# 2. density : True 설정 시, y축이 상대도수 값이 되며 False 설정 시, 데이터의 수로 그래프를 그릴 수 있음
# 3. alpha : 그래프 색상의 투명도를 설정
# 4. histtype : step으로 설정 시, 히스토그램 내부의 색은 비어있는 그래프, stepfilled로 설정 시, 내부를 색으로 채워서 그릴 수 있음
# 5. color : 그래프 색상 설정
# 6. label : 범례 표시 시, 해당 히스토그램의 네이밍 설정
```
